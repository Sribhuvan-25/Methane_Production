```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Assuming your DataFrame is 'df_use'
# Assuming 'ACE-Xi' is the target column
target_column = 'ACE-Xi'

# Remove rows with 'x' in any column
df_cleaned = df_use[~df_use.isin(['x']).any(axis=1)]

# Find the index of the column that starts with "Digester"
digester_col_index = df_cleaned.columns.get_loc("Average-Total-ISD-Cells")

# Combine all columns before "Digester" into a single feature vector
df_cleaned['1D_feature_vector'] = df_cleaned.iloc[:, :digester_col_index].apply(lambda row: row.values, axis=1)

# Retain the columns from "Digester" onwards as separate features (excluding the target column)
other_features = df_cleaned.iloc[:, digester_col_index:-1]  # Excluding target column

# Combine the 1D feature vector and the other features into the final feature set X
X = pd.concat([df_cleaned['1D_feature_vector'], other_features], axis=1)

# Extract the target column
y = df_cleaned[target_column]

# Initialize KFold for cross-validation with 5 splits
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize lists to store scores
mse_scores = []
r2_scores = []

fold = 1

# Perform 5-fold cross-validation
for train_idx, test_idx in kf.split(X):
    print(f"\nTraining fold {fold}...")
    
    # Split the data into training and testing sets
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    # Extract the combined feature vector and convert it into a format the model can use
    X_train_feature_vector = np.stack(X_train['1D_feature_vector'].values)
    X_test_feature_vector = np.stack(X_test['1D_feature_vector'].values)

    # Concatenate the 1D feature vector with other features for training and testing
    X_train_combined = np.concatenate([X_train_feature_vector, X_train.iloc[:, 1:].values], axis=1)
    X_test_combined = np.concatenate([X_test_feature_vector, X_test.iloc[:, 1:].values], axis=1)

    # Initialize the Linear Regression model
    model = LinearRegression()

    # Train the model on the training data
    model.fit(X_train_combined, y_train)

    # Make predictions on the test data
    y_pred = model.predict(X_test_combined)

    # Evaluate the model's performance
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Store the scores for this fold
    mse_scores.append(mse)
    r2_scores.append(r2)

    print(f"Fold {fold} - Mean Squared Error: {mse:.4f}, R-squared: {r2:.4f}")
    fold += 1

# Print average performance across all folds
print("\nCross-validation results:")
print(f"Mean Squared Error: {np.mean(mse_scores):.4f} ± {np.std(mse_scores):.4f}")
print(f"R-squared: {np.mean(r2_scores):.4f} ± {np.std(r2_scores):.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GroupKFold
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten
from tensorflow.keras.optimizers import Adam

# Assuming your DataFrame is 'df_use'
# Assuming 'ACE-Xi' is the target column
target_column = 'ACE-Xi'

# Remove rows with 'x' in any column
df_cleaned = df_use[~df_use.isin(['x']).any(axis=1)]

# Separate the features (X) and the target (y)
X = df_cleaned.drop(columns=[target_column])
y = df_cleaned[target_column].values.astype(np.float32)  # Ensure target is float32

# Extract all digester group columns
digester_columns = [col for col in df_cleaned.columns if col.startswith("Digester")]

# Identify which digester each row belongs to
digester_groups = df_cleaned[digester_columns].idxmax(axis=1).values

# Ensure that all features in X are numeric and of type float32
X = X.astype(np.float32)
X = np.expand_dims(X.values, axis=-1)  # Reshape to be (samples, features, 1) for Conv1D

# Define the model for 1D neural network with 1x1 convolutions
def create_1d_conv_model(input_shape):
    model = Sequential()
    model.add(Conv1D(64, kernel_size=1, activation='relu', input_shape=input_shape))
    model.add(Conv1D(32, kernel_size=1, activation='relu'))
    model.add(Flatten())  # Flatten the output from Conv1D before feeding it to Dense layers
    model.add(Dense(16, activation='relu'))
    model.add(Dense(1, activation='linear'))  # Output layer for regression
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mean_squared_error'])
    return model

# Perform leave-one-digester-group-out cross-validation
unique_digesters = np.unique(digester_groups)
fold = 1

mse_scores = []
r2_scores = []

for digester in unique_digesters:
    print(f"\nTraining fold {fold}, leaving out digester group: {digester}")
    
    # Split the data into training and testing sets based on the current digester group
    train_idx = np.where(digester_groups != digester)[0]
    test_idx = np.where(digester_groups == digester)[0]
    
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    # Ensure that the input data is float32 and properly reshaped
    X_train = X_train.astype(np.float32)
    X_test = X_test.astype(np.float32)

    # Create and train the 1D CNN model
    model = create_1d_conv_model((X_train.shape[1], X_train.shape[2]))  # Adjust input shape
    model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

    # Make predictions on the test data
    y_pred = model.predict(X_test)

    # Evaluate the model's performance
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Store the scores for this fold
    mse_scores.append(mse)
    r2_scores.append(r2)

    print(f"Fold {fold} - Mean Squared Error: {mse:.4f}, R-squared: {r2:.4f}")
    fold += 1

# Print average performance across all folds
print("\nCross-validation results:")
print(f"Mean Squared Error: {np.mean(mse_scores):.4f} ± {np.std(mse_scores):.4f}")
print(f"R-squared: {np.mean(r2_scores):.4f} ± {np.std(r2_scores):.4f}")
```
