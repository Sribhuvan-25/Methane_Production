```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score

# -------------------------
# Common Dataset Definition
# -------------------------
# (Assumes that columns in target_columns are NOT features)
target_columns = [
    "Average-Total-ISD-Cells", "ACE-Xi", "ACE-km", "ACE-Ks", "H2-Xi", "H2-km", "H2-Ks",
    "Digester_BD", "Digester_BF", "Digester_CB", "Digester_CP", "Digester_FD", "Digester_GB",
    "Digester_GP", "Digester_JB", "Digester_LP", "Digester_MA", "Digester_NB", "Digester_NS",
    "Digester_PC", "Digester_PO", "Digester_SF", "Digester_SS", "Digester_SW", "Digester_WA",
    "Digester_WP", "Digester_WR", "Source_I", "Source_M", "Source_P", "Type_CSTR", "Type_EFB",
    "Type_EGSB", "Type_Lagoon", "Type_UASB", "Waste_BW", "Waste_Dairy", "Waste_FW", "Waste_HSI",
    "Waste_MPW", "Waste_MS", "Waste_MS+Dairy", "Waste_MS+HSI", "Waste_PP", "Waste_PR",
    "Waste_SDW", "Biomass_F", "Biomass_G"
]

class TabularDataset(Dataset):
    def __init__(self, df, cat_cols, num_cols, target_col, scaler=None, cat_encoders=None):
        self.df = df.copy().dropna().reset_index(drop=True)
        # Convert to numeric and fill missing values
        self.df = self.df.apply(pd.to_numeric, errors='coerce').fillna(0)
        self.cat_cols = cat_cols
        self.num_cols = num_cols
        self.target_col = target_col
        
        # Scale numerical features
        if scaler is None:
            self.scaler = StandardScaler()
            self.num_data = self.scaler.fit_transform(self.df[num_cols])
        else:
            self.scaler = scaler
            self.num_data = self.scaler.transform(self.df[num_cols])
        
        # Encode categorical features
        self.cat_encoders = {}
        cat_data_list = []
        for col in cat_cols:
            if cat_encoders is None or col not in cat_encoders:
                le = LabelEncoder()
                encoded = le.fit_transform(self.df[col])
                self.cat_encoders[col] = le
            else:
                le = cat_encoders[col]
                self.cat_encoders[col] = le
                encoded = le.transform(self.df[col])
            cat_data_list.append(encoded.astype(np.int64))
        if len(cat_cols) > 0:
            self.cat_data = np.stack(cat_data_list, axis=1)
        else:
            self.cat_data = None
        
        self.target = self.df[target_col].values.astype(np.float32)
        
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        num = self.num_data[idx]
        if self.cat_data is not None:
            cat = self.cat_data[idx]
            return (torch.tensor(num, dtype=torch.float32),
                    torch.tensor(cat, dtype=torch.long),
                    torch.tensor(self.target[idx], dtype=torch.float32))
        else:
            return (torch.tensor(num, dtype=torch.float32),
                    torch.tensor(self.target[idx], dtype=torch.float32))

# -------------------------
# TabTransformer Model
# -------------------------
class TabTransformer(nn.Module):
    def __init__(self, num_features, cat_cardinalities, embed_dim=32, nhead=4, num_layers=2,
                 dim_feedforward=64, dropout=0.1):
        """
        num_features: Number of numerical features.
        cat_cardinalities: List with the number of unique values for each categorical feature.
        """
        super(TabTransformer, self).__init__()
        # Embedding for categorical features
        self.cat_embeddings = nn.ModuleList([
            nn.Embedding(num_embeddings=card, embedding_dim=embed_dim)
            for card in cat_cardinalities
        ])
        # Projection for numerical features (projects entire vector into embed_dim)
        self.num_proj = nn.Linear(num_features, embed_dim)
        
        # We will treat the numerical projection and each categorical embedding as tokens
        self.num_tokens = 1 + len(cat_cardinalities)  # one token for numerical features
        self.transformer_input_dim = self.num_tokens * embed_dim
        
        # Transformer encoder expects shape [seq_len, batch, d_model]
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=self.transformer_input_dim,
            nhead=nhead,
            dropout=dropout,
            dim_feedforward=dim_feedforward
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        
        # Regression head
        self.regressor = nn.Sequential(
            nn.Linear(self.transformer_input_dim, dim_feedforward),
            nn.ReLU(),
            nn.Linear(dim_feedforward, 1)
        )
        
    def forward(self, num_data, cat_data):
        batch_size = num_data.size(0)
        # Project numerical features and view as a token
        num_embed = self.num_proj(num_data)  # [batch, embed_dim]
        tokens = [num_embed]
        # Embed each categorical column (if available)
        if cat_data is not None:
            cat_embeds = [emb(cat_data[:, i]) for i, emb in enumerate(self.cat_embeddings)]
            tokens.extend(cat_embeds)
        # Stack tokens: shape becomes [batch, num_tokens, embed_dim]
        tokens = torch.stack(tokens, dim=1)
        # Flatten tokens for each sample into a single vector
        tokens = tokens.view(batch_size, -1)  # [batch, transformer_input_dim]
        # Add a dummy sequence dimension (seq_len=1) for the transformer
        tokens = tokens.unsqueeze(0)  # [1, batch, transformer_input_dim]
        transformed = self.transformer(tokens)  # [1, batch, transformer_input_dim]
        transformed = transformed.squeeze(0)  # [batch, transformer_input_dim]
        output = self.regressor(transformed)
        return output.squeeze(1)

# -------------------------
# Cross-Validation and Training for TabTransformer
# -------------------------
def cross_validate_tabtransformer(df, target_column, cat_cols, num_cols, k_folds=5,
                                  epochs=50, batch_size=32, learning_rate=0.001):
    kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    mse_scores, r2_scores = [], []
    all_actuals, all_predictions = [], []
    
    for fold, (train_idx, val_idx) in enumerate(kf.split(df)):
        print(f"\n[TabTransformer] Fold {fold+1}/{k_folds}")
        df_train = df.iloc[train_idx].reset_index(drop=True)
        df_val   = df.iloc[val_idx].reset_index(drop=True)
        train_ds = TabularDataset(df_train, cat_cols, num_cols, target_column)
        val_ds = TabularDataset(df_val, cat_cols, num_cols, target_column,
                                scaler=train_ds.scaler, cat_encoders=train_ds.cat_encoders)
        train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
        val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False)
        
        # Determine categorical cardinalities
        cat_cardinalities = [len(train_ds.cat_encoders[col].classes_) for col in cat_cols]
        input_dim = len(num_cols)
        
        model = TabTransformer(num_features=input_dim, cat_cardinalities=cat_cardinalities,
                               embed_dim=32, nhead=4, num_layers=2,
                               dim_feedforward=64, dropout=0.1).to(device)
        criterion = nn.MSELoss()
        optimizer = optim.Adam(model.parameters(), lr=learning_rate)
        
        model.train()
        for epoch in range(epochs):
            epoch_loss = 0
            for batch in train_loader:
                if len(cat_cols) > 0:
                    X_num, X_cat, y = batch
                    X_num, X_cat, y = X_num.to(device), X_cat.to(device), y.to(device)
                    optimizer.zero_grad()
                    preds = model(X_num, X_cat)
                else:
                    X_num, y = batch
                    X_num, y = X_num.to(device), y.to(device)
                    optimizer.zero_grad()
                    preds = model(X_num, None)
                loss = criterion(preds, y)
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()
            print(f"Epoch [{epoch+1}/{epochs}] Loss: {epoch_loss/len(train_loader):.4f}")
        
        model.eval()
        fold_preds, fold_actuals = [], []
        with torch.no_grad():
            for batch in val_loader:
                if len(cat_cols) > 0:
                    X_num, X_cat, y = batch
                    X_num, X_cat, y = X_num.to(device), X_cat.to(device), y.to(device)
                    preds = model(X_num, X_cat)
                else:
                    X_num, y = batch
                    X_num, y = X_num.to(device), y.to(device)
                    preds = model(X_num, None)
                fold_preds.extend(preds.cpu().numpy())
                fold_actuals.extend(y.cpu().numpy())
        mse = mean_squared_error(fold_actuals, fold_preds)
        r2 = r2_score(fold_actuals, fold_preds)
        mse_scores.append(mse)
        r2_scores.append(r2)
        all_actuals.extend(fold_actuals)
        all_predictions.extend(fold_preds)
        print(f"Fold {fold+1} Results: MSE = {mse:.4f}, R² = {r2:.4f}")
    
    print(f"\n[TabTransformer] Final CV: Mean MSE = {np.mean(mse_scores):.4f}, Mean R² = {np.mean(r2_scores):.4f}")
    plt.figure(figsize=(8,6))
    sns.regplot(x=all_actuals, y=all_predictions, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.title(f"TabTransformer: Predicted vs Actual ({target_column})")
    plt.show()

# -------------------------
# Main Execution for TabTransformer
# -------------------------
if __name__ == "__main__":
    df = pd.read_csv("Data/New_data.csv")  # Update path as needed
    target_column = "H2-km"  # Set your target column
    # Define features: all columns not in target_columns
    cat_cols = []  # (Populate if you have categorical features)
    num_cols = [col for col in df.columns if col not in target_columns]
    cross_validate_tabtransformer(df, target_column, cat_cols, num_cols,
                                  k_folds=5, epochs=100, batch_size=32, learning_rate=0.001)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

# Import the TabTransformer from the external package
from tab_transformer_pytorch import TabTransformer

# -------------------------
# Dataset Definition
# -------------------------
class TabularDataset(Dataset):
    def __init__(self, df, cat_cols, num_cols, target_col, cat_encoders=None):
        self.df = df.copy().dropna().reset_index(drop=True)
        # Convert everything to numeric and fill missing values
        self.df = self.df.apply(pd.to_numeric, errors='coerce').fillna(0)
        self.cat_cols = cat_cols
        self.num_cols = num_cols
        self.target_col = target_col

        # Continuous features (leave unscaled; normalization will be handled by the model)
        self.cont_data = self.df[num_cols].values.astype(np.float32)

        # Process categorical features
        if len(cat_cols) > 0:
            self.cat_encoders = {}
            cat_data_list = []
            for col in cat_cols:
                if cat_encoders is None or col not in cat_encoders:
                    le = LabelEncoder()
                    encoded = le.fit_transform(self.df[col])
                    self.cat_encoders[col] = le
                else:
                    le = cat_encoders[col]
                    self.cat_encoders[col] = le
                    encoded = le.transform(self.df[col])
                cat_data_list.append(encoded.astype(np.int64))
            # Shape: (n_samples, n_cat_features)
            self.cat_data = np.stack(cat_data_list, axis=1)
        else:
            # If no categorical features, create an empty array with shape (n_samples, 0)
            self.cat_data = np.empty((len(self.df), 0), dtype=np.int64)
            self.cat_encoders = None

        # Target values
        self.target = self.df[target_col].values.astype(np.float32)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # Return categorical data, continuous data, and the target.
        # (Note: the imported TabTransformer expects categorical input first.)
        cat = torch.tensor(self.cat_data[idx], dtype=torch.long)
        cont = torch.tensor(self.cont_data[idx], dtype=torch.float32)
        target = torch.tensor(self.target[idx], dtype=torch.float32)
        return cat, cont, target

# -------------------------
# Cross-Validation and Training using Imported TabTransformer
# -------------------------
def cross_validate_tabtransformer(df, target_column, cat_cols, num_cols, k_folds=5,
                                  epochs=50, batch_size=32, learning_rate=0.001):
    kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    mse_scores, r2_scores = [], []
    all_actuals, all_predictions = [], []

    for fold, (train_idx, val_idx) in enumerate(kf.split(df)):
        print(f"\n[TabTransformer] Fold {fold+1}/{k_folds}")
        df_train = df.iloc[train_idx].reset_index(drop=True)
        df_val   = df.iloc[val_idx].reset_index(drop=True)

        # Create training and validation datasets.
        train_ds = TabularDataset(df_train, cat_cols, num_cols, target_column)
        val_ds   = TabularDataset(df_val, cat_cols, num_cols, target_column,
                                  cat_encoders=train_ds.cat_encoders)
        train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
        val_loader   = DataLoader(val_ds, batch_size=batch_size, shuffle=False)

        # Determine categorical cardinalities if there are any categorical features
        if len(cat_cols) > 0:
            cat_cardinalities = tuple(
                len(train_ds.cat_encoders[col].classes_) for col in cat_cols
            )
        else:
            cat_cardinalities = tuple()  # empty tuple when no categorical features

        num_continuous = len(num_cols)

        # Compute mean and std for continuous features from the training set (for normalization)
        cont_train = train_ds.cont_data  # shape: (n_samples, num_continuous)
        cont_mean = np.mean(cont_train, axis=0)
        cont_std = np.std(cont_train, axis=0)
        # Avoid division by zero
        cont_std[cont_std == 0] = 1.0
        # Create a tensor of shape (num_continuous, 2) where each row is [mean, std]
        cont_mean_std = torch.tensor(np.stack([cont_mean, cont_std], axis=1), dtype=torch.float32)

        # Create the TabTransformer model (using the imported implementation)
        model = TabTransformer(
            categories = cat_cardinalities,
            num_continuous = num_continuous,
            dim = 32,
            dim_out = 1,
            depth = 6,
            heads = 8,
            attn_dropout = 0.1,
            ff_dropout = 0.1,
            mlp_hidden_mults = (4, 2),
            mlp_act = nn.ReLU(),
            continuous_mean_std = cont_mean_std
        ).to(device)

        criterion = nn.MSELoss()
        optimizer = optim.Adam(model.parameters(), lr=learning_rate)

        # Training loop for this fold
        model.train()
        for epoch in range(epochs):
            epoch_loss = 0
            for batch in train_loader:
                x_cat, x_cont, y = batch
                x_cat = x_cat.to(device)
                x_cont = x_cont.to(device)
                y = y.to(device)
                optimizer.zero_grad()
                # Even if there are no categorical features, x_cat is an empty tensor with shape (batch_size, 0)
                preds = model(x_cat, x_cont)
                loss = criterion(preds, y)
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()
            print(f"Epoch [{epoch+1}/{epochs}] Loss: {epoch_loss/len(train_loader):.4f}")

        # Evaluation on the validation set for this fold
        model.eval()
        fold_preds, fold_actuals = [], []
        with torch.no_grad():
            for batch in val_loader:
                x_cat, x_cont, y = batch
                x_cat = x_cat.to(device)
                x_cont = x_cont.to(device)
                y = y.to(device)
                preds = model(x_cat, x_cont)
                fold_preds.extend(preds.cpu().numpy())
                fold_actuals.extend(y.cpu().numpy())
        mse = mean_squared_error(fold_actuals, fold_preds)
        r2 = r2_score(fold_actuals, fold_preds)
        mse_scores.append(mse)
        r2_scores.append(r2)
        all_actuals.extend(fold_actuals)
        all_predictions.extend(fold_preds)
        print(f"Fold {fold+1} Results: MSE = {mse:.4f}, R² = {r2:.4f}")

    print(f"\n[TabTransformer] Final CV: Mean MSE = {np.mean(mse_scores):.4f}, Mean R² = {np.mean(r2_scores):.4f}")
    plt.figure(figsize=(8,6))
    sns.regplot(x=all_actuals, y=all_predictions, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.title(f"TabTransformer: Predicted vs Actual ({target_column})")
    plt.show()

# -------------------------
# Main Execution
# -------------------------
if __name__ == "__main__":
    # Load your data (update the CSV path as needed)
    df = pd.read_csv("Data/New_data.csv", index_col=0)
    target_column = "ACE-km"  # set your target column

    # Define feature columns:
    # (For categorical features, list their column names; for continuous features, use all columns not in target_columns.)
    cat_cols = []  # <-- Populate this list if you have categorical features
    target_columns = [
        "Average-Total-ISD-Cells", "ACE-Xi", "ACE-km", "ACE-Ks", "H2-Xi", "H2-km", "H2-Ks",
        "Digester_BD", "Digester_BF", "Digester_CB", "Digester_CP", "Digester_FD", "Digester_GB",
        "Digester_GP", "Digester_JB", "Digester_LP", "Digester_MA", "Digester_NB", "Digester_NS",
        "Digester_PC", "Digester_PO", "Digester_SF", "Digester_SS", "Digester_SW", "Digester_WA",
        "Digester_WP", "Digester_WR", "Source_I", "Source_M", "Source_P", "Type_CSTR", "Type_EFB",
        "Type_EGSB", "Type_Lagoon", "Type_UASB", "Waste_BW", "Waste_Dairy", "Waste_FW", "Waste_HSI",
        "Waste_MPW", "Waste_MS", "Waste_MS+Dairy", "Waste_MS+HSI", "Waste_PP", "Waste_PR",
        "Waste_SDW", "Biomass_F", "Biomass_G"
    ]
    num_cols = [col for col in df.columns if col not in target_columns]

    # Run cross-validation and training
    cross_validate_tabtransformer(df, target_column, cat_cols, num_cols,
                                  k_folds=5, epochs=100, batch_size=32, learning_rate=0.001)
```
